```python
#Welcome To Two-Player Hangman
from IPython.display import clear_output
from Hangman import *                                         #Import every method from Hangman class
import random

print("Welcome to Hangman! \n\nPlease select how many players will be playing below.\n")

players = int(input('Will there be "1" or "2" players: '))    #Choose how many players will playing
print()

difficulty = input('Will you be playing on "easy", "medium", or "hard" difficulty: ').lower()  #Choose which word list to use
print()

if players == 1:
    game = Hangman()
    
    if difficulty == "easy":
        word_list = game.easy_list()    #Create Hangman object and the word list
    elif difficulty == "medium":
        word_list = game.medium_list()
    else:
        word_list = game.hard_list()
        
    word = random.choice(word_list).lower()  #Randomly choose a word from the list
    score = len(word)                         #Get the length of the word to calculate score later on
    place_holder = game.place_holder(word)    #Create the visual place holder display for the player
    print("Your word is: " + place_holder + "\n")
        
    lives = 6
    keeping_display = []                     #Initialize player lives and characters already in the word to correctly modify display
    guessed_letters = []
        
    while lives != 0:
        clear_output(wait=True)
        guess = input("Take a guess: ")
        print()
        guessed_letters.append(guess)
        display = ""
            
        for char in word:
            if char == guess:
                display += char + " "
                keeping_display.append(guess)
            elif char in keeping_display:
                display += char + " "
            else:
                display += "_ " 
                    
        if guess in display:
            lives = lives
        else:
            lives -= 1
            
        guessed_letters.sort()        
        print(f'Your word is: {display} Remaining lives: {lives}/6 Letters guessed: {guessed_letters}\n')
        print(game.stages(lives))
            
        if "_" not in display:
            print(f'Congratulations you guessed the word!\nYour final score is: {lives * score}')
            break
        else:
            continue
    print(f'You lost all your lives.\nYour word was: {word}')
else:
    game = Hangman()
    
    if difficulty == "easy":
        word_list = game.easy_list()    #Create Hangman object and the word list
    elif difficulty == "medium":
        word_list = game.medium_list()
    else:
        word_list = game.hard_list()
        
    word = random.choice(word_list).lower()  #Randomly choose a word from the list
    score = len(word)                         #Get the length of the word to calculate score later on
    place_holder = game.place_holder(word)    #Create the visual place holder display for the player
        
    lives = 6
    keeping_display = []                     #Initialize player lives and characters already in the word to correctly modify display
    guessed_letters = []
    
    player_1_name = input("Player 1, please enter your name to get started: ")
    print()
    
    print(f'{player_1_name}, your word is: {place_holder}\n')
    
    while lives != 0:
        clear_output(wait=True)
        guess = input("Take a guess: ")
        print()
        guessed_letters.append(guess)
        display = ""
            
        for char in word:
            if char == guess:
                display += char + " "
                keeping_display.append(guess)
            elif char in keeping_display:
                display += char + " "
            else:
                display += "_ " 
                    
        if guess in display:
            lives = lives
        else:
            lives -= 1
            
        guessed_letters.sort()        
        print(f'Your word is: {display} Remaining lives: {lives}/6 Letters guessed: {guessed_letters}\n')
        print(game.stages(lives))
            
        if "_" not in display:
            player_1_score = lives * score
            print(f'Congratulations you guessed the word!\nYour final score is: {player_1_score}\n')
            break
        else:
            continue
    player_1_score = lives * score

    if lives == 0:
        print(f'You lost all your lives.\nYour word was: {word}\n')
    
    word = random.choice(word_list).lower()  #Randomly choose a word from the list
    score = len(word)                         #Get the length of the word to calculate score later on
    place_holder = game.place_holder(word)    #Create the visual place holder display for the player
        
    lives = 6
    keeping_display = []                     #Initialize player lives and characters already in the word to correctly modify display
    guessed_letters = []
    
    player_2_name = input("Player 2, please enter your name to get started: ")
    print()
    
    print(f'{player_2_name}, your word is: {place_holder}\n')
    
    while lives != 0:
        clear_output(wait=True)
        guess = input("Take a guess: ")
        print()
        guessed_letters.append(guess)
        display = ""
            
        for char in word:
            if char == guess:
                display += char + " "
                keeping_display.append(guess)
            elif char in keeping_display:
                display += char + " "
            else:
                display += "_ " 
                    
        if guess in display:
            lives = lives
        else:
            lives -= 1
            
        guessed_letters.sort()        
        print(f'Your word is: {display} Remaining lives: {lives}/6 Letters guessed: {guessed_letters}\n')
        print(game.stages(lives))
            
        if "_" not in display:
            player_2_score = lives * score
            print(f'Congratulations you guessed the word!\nYour final score is: {player_2_score}')
            break
        else:
            continue
            
    if lives == 0:
        print(f'You lost all your lives.\nYour word was: {word}')
    player_2_score = lives * score
        
    if player_1_score > player_2_score:
        print()
        print(f'{player_1_name} smoked {player_2_name}!\nWith the final score being {player_1_score} to {player_2_score}')
    elif player_1_score < player_2_score:
        print()
        print(f'{player_2_name} smoked {player_1_name}!\nWith the final score being {player_2_score} to {player_1_score}')
    else:
        print()
        print(f'You will have to play again because it was a tie!\nWith the final score being {player_1_score} to {player_2_score}')
```
